```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import pandas as pd

import pickle
```

```python
print("Loading pretrained InceptionV1 weights...")
file = open('pretrained_models/places-googlenet.pickle', "rb")
weights = pickle.load(file, encoding="bytes")

# Extracting keys and shapes
data = {"Key": [], "Shape": []}
for key, value in weights.items():
    data["Key"].append(key.decode("utf-8"))  # Decoding bytes to string
    data["Shape"].append(value.shape)

# Creating DataFrame
df = pd.DataFrame(data)

# Setting display options
pd.set_option('display.max_rows', None)  # Show all rows
pd.set_option('display.max_columns', None)  # Show all columns
pd.set_option('display.width', None)  # Auto-detect the display width
pd.set_option('display.max_colwidth', None)  # Display full width of columns

# Displaying the DataFrame
print(df)

file.close()
```
